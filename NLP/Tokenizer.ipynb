```python
corpus="""My name is Harshit. I am learning NLP! and lets goo we will do it."""
```

```python
print(corpus)
```

```python
from nltk.tokenize import sent_tokenize
```

```python
sentence=sent_tokenize(corpus)
```

```python
print(sentence)
```

```python
for sent in sentence:
    print(sent)
```

```python
from nltk.tokenize import word_tokenize
word_tokenize(corpus)
```

```python
for sent in sentence:
    print(word_tokenize(sent))
```

```python
from nltk.tokenize import wordpunct_tokenize
wordpunct_tokenize(corpus)
```

Stemming - USE for classification Problems

```python
#porterStemming
words=['eat','eating','eaten','writing','writer','programming','programs','history','finally','finalized']
from nltk.stem import PorterStemmer
ps=PorterStemmer()
for word in words:
    print(word+'------>'+ps.stem(word))
```

```python
#RegxStemmer 
from nltk.stem import RegexpStemmer
stemmer=RegexpStemmer('ing$|s$|e$|able$',min=4)
print(stemmer.stem('eating'))
print(stemmer.stem('eats'))
print(stemmer.stem('writing'))
```

```python
#snowball stemmer-> best among all of them
from nltk.stem import SnowballStemmer
snowball=SnowballStemmer('english')
for word in words:
    print(word+'---->'+snowball.stem(word))
```

```python
#comparing porter and snowball
#porter
ps.stem('fairly'),ps.stem('sportingly')
```

```python
#snowball
snowball.stem('fairly'),snowball.stem('sportingly')
```

Lemmatization NLTK

```python
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
'''pos
    verb-v,
    noun-n,
    adverb-a,
    adjective-r
'''
print(lemmatizer.lemmatize('going',pos='v'))
print(lemmatizer.lemmatize('going',pos='n'))
```

```python
for word in words:
    print(word+'---------->'+lemmatizer.lemmatize(word,pos='v'))
```
