```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('Algerian_forest_fires_dataset.csv')
df.head()
```

```python
df.info()
```

```python
df.isnull().any(axis=1)
```

```python
df[(df.isnull().any(axis=1))]
```

```python
df.shape
```

```python
df=df.dropna().reset_index(drop=True)
```

```python
df[(df.isnull().any(axis=1))]
```

```python
df.iloc[122]
```

```python
df.shape
```

```python
#drop 122 row
df = df.drop(df.index[122]).reset_index(drop=True)
```

```python
df.shape
```

```python
df.loc[:122,'Region']=0
df.loc[122:,'Region']=1
```

```python
df.columns
```

```python
df.columns=df.columns.str.strip( )
```

```python
df.columns
```

```python
df.head()
```

```python
df[['day','month','year','Temperature','RH','Ws']]=df[['day','month','year','Temperature','RH','Ws']].astype(int)
```

```python
df.info()
```

```python
df.head()
```

```python
features=[]
for col in df.columns:
    if df[col].dtype=='O':
        features.append(col)
features
```

```python
for i in features:
    if i!='Classes':
        df[i]=df[i].astype(float)
```

```python
df.info()
```

```python
df.to_csv('Algerian_forest_fires_dataset_cleaned.csv',index=False)
```

```python
df['Classes'].value_counts()
```

```python
df['Classes']=np.where(df['Classes'].str.contains('not fire'),0,1)
```

```python
df['Classes'].value_counts()
```

```python
df_copy=df
```

```python
df_copy=df_copy.drop(['day','month','year'],axis=1)
```

```python
df_copy.head()
```

```python
#pie chart
percentage=df_copy['Classes'].value_counts(normalize=True)*100
percentage
```

```python
#pie chart 
plt.figure(figsize=(10,5))
label=['fire','not fire']
color=['green','red']
plt.pie(percentage,labels=label,colors=color, autopct='%1.1f%%', startangle=90)
plt.show()
```

```python
#monthly analysis
dftemp=df.loc[df['Region']==1]
plt.figure(figsize=(5,5))
sns.set_style('whitegrid')
sns.countplot(x='month',hue='Classes',data=dftemp)

plt.title('Fire vs non fire (Region=1)')
plt.xlabel('months')
plt.ylabel('count')
plt.show()
```

```python
#monthly analysis
dftemp=df.loc[df['Region']==0]
plt.figure(figsize=(5,5))
sns.set_style('whitegrid')
sns.countplot(x='month',hue='Classes',data=dftemp)

plt.title('Fire vs non fire (Region 0)')
plt.xlabel('months')
plt.ylabel('count')
plt.show()
```
